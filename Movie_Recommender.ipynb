**Introduction**

A movie recommender system is a sophisticated machine learning application designed to help users navigate vast film libraries by suggesting content tailored to their individual tastes.

The system primarily operates using two core strategies:

**Collaborative-Based Filtering:** This method recommends movies based on the viewing habits of similar users ("people like you also watched..."). It identifies patterns in ratings, watch history, and user interactions to make predictions.

**Content-Based Filtering:** This method recommends movies that are similar in features (genre, director, actors, keywords, etc.) to films the user has previously liked.

**Converting Dictionary into lists**

```python
import ast
ast.literal_eval
```

```python
import ast

def convert(obj):
  l=[]


  if isinstance(obj, list):
    return obj
  try:
    for i in ast.literal_eval(obj):
      l.append(i['name'])
  except (ValueError, SyntaxError):
    # Handle cases where the string is not a valid literal (e.g., empty string or other formats)
    pass
  return l
```

**Converting genres into a list**

```python
movies['genres']=movies['genres'].apply(convert)
```

**Checking for the genre column**

```python
movies.head()
```

**Converting keywords into a list**

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

**Converting cast into a list**

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

**Converting crew into a list**

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

**Stemming the unnnecessary words**

```python
import nltk
```

```python
pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

**Creating a recommend movie function**

**Model Building**

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)

```

**Model Testing**

```python
recommend("Avatar")
```
